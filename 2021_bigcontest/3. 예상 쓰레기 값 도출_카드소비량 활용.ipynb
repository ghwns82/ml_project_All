```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv("csv/카드소비량 분석.csv")
```

```python
df["Month"] = df["base_date"].str[-2:]
df["Year"] = df["base_date"].str[:4]
df
```

# 각 기울기(변화량) 값 구하기

```python
food_list= list(df.columns)[3:-2]

기울기_2020_7 = []
for food in food_list:
    기울기_2020_7.append(df[df['base_date'] == '2020-07'][food].values - df[df['base_date'] == '2020-06'][food].values)
기울기_2020_8 = []
for food in food_list:
    기울기_2020_8.append(df[df['base_date'] == '2020-08'][food].values - df[df['base_date'] == '2020-06'][food].values)
 
    
len(기울기_2020_7), len(기울기_2020_7[0]), len(기울기_2020_8), len(기울기_2020_8[0])
```

```python
half_month_index=df[(df['Month'] == '07') |  (df['Month'] == '08') |  (df['Month'] == '09') |  (df['Month'] == '10') |  (df['Month'] == '11')|  (df['Month'] == '12')].index
half_y_rate_df=df.drop(half_month_index)
half_y_rate_df
```

```python
# 각 년도별 1~6월까지의 합
year_df = half_y_rate_df.groupby(['Year','emd_nm']).sum()
year_df= year_df.reset_index()
year_df 
```

## 인덱스리스트와 food리스트로 각 값을 추출하자

### 7월

```python
index_list_2021= list(year_df[year_df['Year']=='2021'].index)
index_list_2020= list(year_df[year_df['Year']=='2020'].index)
year_df_2021 = year_df[year_df['Year']=='2021']
year_df_2020 = year_df[year_df['Year']=='2020']
```

```python
최종_변화량 = []
for f, food in enumerate(food_list):
    음식별_동예측리스트=[]
    for i, index in enumerate(index_list_2020):
        음식별_동예측리스트.append(기울기_2020_7[f][i] * (year_df_2021.loc[index_list_2021[i], food]/year_df_2020.loc[index_list_2020[i], food]))
    최종_변화량.append(음식별_동예측리스트)
```

```python
len(최종_변화량),len(최종_변화량[0]), 최종_변화량[0][0]
```

```python
예상_7월 = df.copy()
예상_7월 = 예상_7월[예상_7월["base_date"]=="2021-06"]
for i, index in enumerate(예상_7월.index):
    for f, food in enumerate(food_list):
        예상_7월.loc[index,food]  =   max(예상_7월.loc[index,food] + 최종_변화량[f][i], 0)
예상_7월=예상_7월.drop(['Year','Month'],axis=1)
예상_7월["base_date"]="2021-07"
예상_7월
```

```python
최종_변화량 = []
for f, food in enumerate(food_list):
    음식별_동예측리스트=[]
    for i, index in enumerate(index_list_2020):
        음식별_동예측리스트.append(기울기_2020_8[f][i] * (year_df_2021.loc[index_list_2021[i], food]/year_df_2020.loc[index_list_2020[i], food]))
    최종_변화량.append(음식별_동예측리스트)
```
