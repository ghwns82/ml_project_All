## 내국인 유동인구 데이터를 거주인구, 근무인구, 방문인구로 나누어봄

## 1.내국인거주인구

```python
from fbprophet.plot import add_changepoints_to_plot


내국인거주인구_df=df_2_1.groupby(['base_date'])['resd_pop_cnt'].mean()*36
내국인거주인구_df=내국인거주인구_df.reset_index()
내국인거주인구_df.columns=['ds','y']
m=Prophet(daily_seasonality=True)
m.fit(내국인거주인구_df)
#지정된 날짜수만큼 미래로 확장된 데이터프레임을 만듦
future=m.make_future_dataframe(periods=62)
pre_future=m.predict(future)
fig=m.plot(pre_future)
a=add_changepoints_to_plot(fig.gca(), m, pre_future)
```

```python
fig = m.plot_components(pre_future)
```
