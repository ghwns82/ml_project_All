# 유동인구 데이터 전처리

```python
df_2_1["base_date"]=pd.to_datetime(df_2_1['base_date'])
df_2_1['Year']=df_2_1['base_date'].dt.year
df_2_1['Month']=df_2_1['base_date'].dt.month
df_2_2["base_date"]=pd.to_datetime(df_2_2['base_date'])
df_2_2['Year']=df_2_2['base_date'].dt.year
df_2_2['Month']=df_2_2['base_date'].dt.month
df_2_3["base_date"]=pd.to_datetime(df_2_3['base_date'])
df_2_3['Year']=df_2_3['base_date'].dt.year
df_2_3['Month']=df_2_3['base_date'].dt.month
```

```python
len(df_2_1['emd_nm'].unique())
```

```python
내국인거주=pd.DataFrame(df_2_1.groupby(['Year','Month'])['resd_pop_cnt'].mean())
내국인근무=pd.DataFrame(df_2_1.groupby(['Year','Month'])['work_pop_cnt'].mean())
내국인방문=pd.DataFrame(df_2_1.groupby(['Year','Month'])['visit_pop_cnt'].mean())
```

```python
장기거주=pd.DataFrame(df_2_2.groupby(['Year','Month'])['resd_pop_cnt'].mean())
장기근무=pd.DataFrame(df_2_2.groupby(['Year','Month'])['work_pop_cnt'].mean())
장기방문=pd.DataFrame(df_2_2.groupby(['Year','Month'])['visit_pop_cnt'].mean())
```

```python
단기방문=pd.DataFrame(df_2_3.groupby(['Year','Month'])['visit_pop_cnt'].mean())
```

```python
내국인거주.reset_index(inplace=True)
내국인근무.reset_index(inplace=True)
내국인방문.reset_index(inplace=True)
장기거주.reset_index(inplace=True)
장기근무.reset_index(inplace=True)
장기방문.reset_index(inplace=True)
단기방문.reset_index(inplace=True)
```

```python
거주=pd.DataFrame(pd.concat([내국인거주,장기거주]).groupby(['Year','Month'])['resd_pop_cnt'].sum())
근무=pd.DataFrame(pd.concat([내국인근무,장기근무]).groupby(['Year','Month'])['work_pop_cnt'].sum())
방문=pd.DataFrame(pd.concat([내국인방문,장기방문,단기방문]).groupby(['Year','Month'])['visit_pop_cnt'].sum())
```

```python
df=pd.concat([거주,근무,방문],axis=1)
df
```

```python
df.reset_index(inplace=True)
df=df.iloc[27:42]
df=df.set_index(['Year','Month'])
df
```
